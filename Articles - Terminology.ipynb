# Terminology analysis
This notebook uses the article.json file to count the terms necessary for the terminology analysis in our project. Below we load in the cell and correct the type of certain columns.

```python
article = pd.read_json('articles.json')
article.publicationDay = pd.to_datetime(article.publicationDay, unit='ms')
```

```python
article.head()
```

```python
data = pd.DataFrame()
data['article'] = article.id
data['text'] = article.bodyText
data['wordcount'] = article.wordcount
data['publicationDay'] = article.publicationDay
data['coronavirusCount'] = data.text.str.lower().str.count('coronavirus')
data['covid19Count'] = data.text.str.lower().str.count('covid-19')
data['sarscov2Count'] = data.text.str.lower().str.count('sars-cov-2')
```

```python
data.to_json('articles_terms.json')
```
