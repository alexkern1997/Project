```python
import pandas as pd
import numpy as np
import glob
pd.set_option('display.max_columns', None)
```

```python
path_to_json = 'tempdata/articles/*.json'
json_file_list = glob.glob(path_to_json)
```

```python
data = pd.DataFrame()
for json in json_file_list:
    temp = pd.read_json(json)
    data = data.append(temp, ignore_index=True)
```

```python
data2 = pd.json_normalize(data.fields)
data = pd.merge(data, data2, left_index=True, right_index=True)
```

```python
# remember if we ever want to split the keywords variable use this function
# pd.DataFrame(data.keywords.values.tolist()).add_prefix('keyword_')
```

```python
keywords=[]
for i in range(len(data)):
    df = pd.json_normalize(data.tags[i])
    keywords.append(df['id'][df['type']=='keyword'].tolist())
data['keywords'] = keywords
```

```python
article_ids=[]
article_keywords = []
for i in range(len(data)):
    article_id= data.id[i]
    for keyword in data.keywords[i]:
        article_ids.append(article_id)
        article_keywords.append(keyword)
```

```python
id_keyword = pd.DataFrame(list(zip(article_ids, article_keywords)), columns=['id','keyword'])
```
